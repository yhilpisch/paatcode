### Notebook Goals

This notebook runs the EMH-style Monte Carlo experiment from `ch21_emh_ai_retail.py` and explores the distribution of random-strategy Sharpe ratios.

- Simulate Sharpe ratios for many edge-free strategies.
- Summarise the distribution numerically and visually.
- Connect the results to multiple-testing concerns in research workflows.


```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")  # set notebook-wide plotting style
try:
    from matplotlib_inline.backend_inline import set_matplotlib_formats
    set_matplotlib_formats("svg")
except Exception:
    pass
plt.rcParams.update({"figure.dpi": 300})

PROJECT_ROOT = Path("..").resolve()  # project root relative to notebooks
CODE_DIR = PROJECT_ROOT / "code"  # folder with chapter scripts
if str(CODE_DIR) not in sys.path:  # ensure scripts are importable
    sys.path.insert(0, str(CODE_DIR))

import ch21_emh_ai_retail as ch21  # EMH Monte Carlo helpers
```

### 1. Monte Carlo Simulation

We use the default configuration from the script and simulate many random strategies with no true edge.

```python
cfg = ch21.EmhSimulationConfig(n_strategies=3_000, n_days=750)
sharpes = ch21.simulate_random_sharpes(cfg)
summary = ch21.sharpe_summary(sharpes)
summary
```

### 2. Histogram of Random Sharpe Ratios

The histogram highlights how easy it is to obtain apparently attractive Sharpe ratios from purely random trading rules once many strategies are tried.

```python
fig, ax = plt.subplots(figsize=(6.0, 3.4))
ax.hist(sharpes, bins=40, color="tab:blue", alpha=0.7, edgecolor="white")
ax.set_xlabel("annualised Sharpe ratio")
ax.set_ylabel("number of strategies")
ax.set_title("Random-strategy Sharpe distribution under EMH baseline")
ax.grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

### Takeaways

- Even when the true expected return is zero, some strategies will look impressive in finite samples.
- The more models and parameter grids you explore, the more you must adjust intuition for what constitutes convincing evidence.
- EMH remains a useful benchmark once these multiple-testing effects are taken seriously in research and deployment decisions.
