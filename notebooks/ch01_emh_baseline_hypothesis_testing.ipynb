### Notebook Goals

This notebook turns the Efficient Market Hypothesis (EMH) into concrete Python objects.

- Build a minimal random-walk laboratory that matches the Chapter 1 discussion.
- Visualise price paths and return distributions under an EMH-style baseline.
- Connect these diagnostics to the hypothesis-testing workflow used throughout the class.


```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")  # set notebook-wide plotting style
try:
    from matplotlib_inline.backend_inline import set_matplotlib_formats
    set_matplotlib_formats("svg")
except Exception:
    pass
plt.rcParams.update({"figure.dpi": 300})

PROJECT_ROOT = Path("..").resolve()  # project root relative to notebooks
CODE_DIR = PROJECT_ROOT / "code"  # folder with standalone chapter scripts
if str(CODE_DIR) not in sys.path:  # ensure scripts are importable
    sys.path.insert(0, str(CODE_DIR))

print(f"Project root: {PROJECT_ROOT}")  # quick sanity check
print(f"Code directory: {CODE_DIR}")  # where chapter scripts live
```

### Random-Walk Laboratory

We start by reusing the simulation helper from the Chapter 2 script `ch02_random_walk_baseline.py`. The goal is to see how an EMH-style baseline behaves before we ever look at real data.

```python
import ch02_random_walk_baseline as rw  # random-walk helpers for EMH

prices, log_ret = rw.simulate_log_random_walk(steps=252 * 4)  # four years

fig, (ax_price, ax_ret) = plt.subplots(2, 1, figsize=(8.0, 4.6))

ax_price.plot(prices)  # simulated price path under EMH baseline
ax_price.set_ylabel("price level")
ax_price.set_title("Random-walk price path under EMH baseline")
ax_price.grid(True, alpha=0.3)

ax_ret.hist(log_ret, bins=40, alpha=0.8, color="tab:blue")
ax_ret.set_xlabel("log-return")
ax_ret.set_ylabel("frequency")
ax_ret.set_title("Distribution of daily log-returns")
ax_ret.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()
```

### Diagnostics as Hypothesis-Testing Ingredients

The same helper script also computes simple diagnostics such as the t-statistic for the mean and short-lag autocorrelations. These numbers will later play the role of summary evidence when we compare strategies to EMH-style baselines.

```python
diag = rw.compute_diagnostics(log_ret, max_lag=5)  # summary numbers
for key, value in diag.items():
    print(f"{key:>8s} : {value: .4f}")
```

### Takeaways

- EMH-style random walks give you a concrete picture of what “no edge” looks like in code and figures.
- Basic diagnostics (mean, volatility, autocorrelations) become building blocks for later hypothesis tests.
- Treat every future strategy as a deviation from this baseline that has to justify itself with evidence.
