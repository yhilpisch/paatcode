### Notebook Goals

This notebook showcases how NumPy and Matplotlib work together in the toy equity example for Chapter 4 and extends it with a few extra diagnostics.

- Reuse the `build_equity_curve` helper from the chapter script.
- Visualise the equity curve and the underlying daily returns.
- Highlight array operations that appear repeatedly in later chapters.


```python
import sys
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")  # set notebook-wide plotting style
try:
    from matplotlib_inline.backend_inline import set_matplotlib_formats
    set_matplotlib_formats("svg")
except Exception:
    pass
plt.rcParams.update({"figure.dpi": 300})

PROJECT_ROOT = Path("..").resolve()  # project root relative to notebooks
CODE_DIR = PROJECT_ROOT / "code"  # folder with chapter scripts
if str(CODE_DIR) not in sys.path:  # ensure scripts are importable
    sys.path.insert(0, str(CODE_DIR))

import ch04_numpy_matplotlib_equity as ch04  # toy equity helpers
```

### 1. Building the Equity Curve

The chapter script defines a tiny vector of daily simple returns and turns it into a cumulative wealth index. We replicate that here and look at the raw numbers.

```python
days, equity = ch04.build_equity_curve()
daily_ret = np.diff(equity) / equity[:-1]  # simple returns implied by path

print("Equity curve values:")
print(equity)
print("\nImplied daily returns:")
print(daily_ret)
```

### 2. Visualising Prices and Returns

We now adapt the original figure to show both the equity curve and the bar chart of daily returns in a single canvas.

```python
fig, (ax_eq, ax_ret) = plt.subplots(2, 1, figsize=(7.0, 4.4))

ax_eq.plot(days, equity, marker="o")  # equity curve line with markers
ax_eq.set_xlabel("time step")
ax_eq.set_ylabel("wealth (normalised)")
ax_eq.set_title("Toy equity curve from NumPy returns")
ax_eq.grid(True, alpha=0.3)

ax_ret.bar(days[1:], daily_ret, width=0.6, color="tab:orange")
ax_ret.set_xlabel("time step")
ax_ret.set_ylabel("daily simple return")
ax_ret.set_title("Daily returns implied by equity curve")
ax_ret.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()
```

### Takeaways

- Small, fully controlled examples are ideal for practising NumPy array manipulations.
- Splitting a figure into multiple panels makes it easier to relate equity curves to the underlying daily returns.
- The same patterns reappear when we work with real EOD or intraday data in later chapters.
