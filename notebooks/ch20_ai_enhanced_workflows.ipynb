### 1. Build and Render a Template

We start with the backtest-review template and render it using example metrics. In practice, these values would come from your own backtests.

```python
template = ch20.build_backtest_review_template()
context = {
    "ann_return": "0.15",
    "ann_vol": "0.20",
    "max_dd": "-0.25",
    "sharpe": "0.75",
    "notes": "Cross-sectional momentum strategy on liquid equities.",
}
prompt_text = template.render(context)
print(prompt_text.splitlines()[0])
print("...")
```

### 2. Demo Prompt Session
