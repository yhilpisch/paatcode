### Notebook Goals

This notebook mirrors the Chapter 3 discussion on platforms and project skeletons with concrete Python checks.

- Inspect the local environment (Python version, paths, key directories).
- Explore how the `code/`, `figures/`, `slides/`, and `notebooks/` folders relate.
- Provide small helper functions that future notebooks can reuse to locate assets.


```python
import sys
from pathlib import Path
import subprocess

import platform
if "google.colab" in sys.modules:
    REPO_URL = "https://github.com/yhilpisch/paatcode.git"
    ROOT_DIR = Path("/content/paatcode")
    if not ROOT_DIR.exists():
        subprocess.run([
            "git",
            "clone",
            REPO_URL,
            str(ROOT_DIR),
        ], check=True)
    PROJECT_ROOT = ROOT_DIR
else:
    PROJECT_ROOT = Path("..").resolve()  # project root relative to notebooks
CODE_DIR = PROJECT_ROOT / "code"  # location of Python scripts
FIG_DIR = PROJECT_ROOT / "figures"  # location of generated figures
SLIDES_DIR = PROJECT_ROOT / "slides"  # location of slide decks
NOTEBOOKS_DIR = PROJECT_ROOT / "notebooks"  # location of notebooks

print(f"Python version : {platform.python_version()}")  # interpreter info
print(f"Project root   : {PROJECT_ROOT}")  # base directory
print(f"Code dir       : {CODE_DIR}")  # scripts and examples
print(f"Figures dir    : {FIG_DIR}")  # diagnostic figures
print(f"Slides dir     : {SLIDES_DIR}")  # slide decks
print(f"Notebooks dir  : {NOTEBOOKS_DIR}")  # this folder

if str(CODE_DIR) not in sys.path:  # enable imports from code directory
    sys.path.insert(0, str(CODE_DIR))
```

### Helper: Resolving Project Paths

The function below gives later notebooks a central way to resolve where assets live. It is deliberately small and avoids any external configuration systems.

```python
def resolve_path(*parts: str) -> Path:
    """Resolve a path inside the project root.

    Parameters
    ----------
    *parts:
        Additional path segments such as ("figures", "ch02_...").
    """
    return PROJECT_ROOT.joinpath(*parts)  # joined and normalised path


resolve_path("code", "ch05_eod_engineering.py")
```

### Visual Map of the Skeleton

Finally, a small text-based map summarises how the main folders relate to the research workflow described in the chapter.

```python
structure = {
    "code": "chapter scripts, wrappers, and streaming helpers",
    "figures": "exported diagnostics and visuals",
    "slides": "beamer slide decks for the class",
    "notebooks": "chapter walk-throughs and experiments",
}

for name, desc in structure.items():
    path = resolve_path(name)
    exists = "yes" if path.exists() else "no"
    print(f"{name:10s} → {path} (exists: {exists})  # {desc}")
```

### Takeaways

- A consistent project skeleton makes it easier to move between scripts, figures, slides, and notebooks.
- Small helpers such as `resolve_path` keep later examples readable while avoiding hard-coded directories.
- Before running more advanced code, always make sure you know where inputs and outputs live on your system.
