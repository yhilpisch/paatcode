### Notebook Goals

This notebook walks through the core scripts for Chapter 2 and extends them with side-by-side visuals.

- Reuse the random-walk baseline to calibrate expectations under EMH.
- Load SPY end-of-day data and compare its diagnostics to the baseline.
- Visualise how close empirical behaviour is to the simulated reference.


```python
import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")  # set notebook-wide plotting style
try:
    from matplotlib_inline.backend_inline import set_matplotlib_formats
    set_matplotlib_formats("svg")
except Exception:
    pass
plt.rcParams.update({"figure.dpi": 300})

PROJECT_ROOT = Path("..").resolve()  # project root relative to notebooks
CODE_DIR = PROJECT_ROOT / "code"  # folder with standalone chapter scripts
if str(CODE_DIR) not in sys.path:  # ensure scripts are importable
    sys.path.insert(0, str(CODE_DIR))

import ch02_random_walk_baseline as rw  # random-walk experiments
import ch02_spy_eod_diagnostics as spy_diag  # SPY diagnostics
```

### 1. Random-Walk Baseline

We first simulate an efficient market with unpredictable shocks and inspect its diagnostics.

```python
prices_rw, log_ret_rw = rw.simulate_log_random_walk(steps=252 * 5)
diag_rw = rw.compute_diagnostics(log_ret_rw, max_lag=5)

print("Random-walk baseline diagnostics")
for key, value in diag_rw.items():
    print(f"{key:>8s} : {value: .4f}")
```

### 2. SPY End-of-Day Data

Next we load real SPY end-of-day prices and compute the same style of diagnostics.

```python
spy_prices = spy_diag.load_spy_prices()  # SPY closing prices
spy_log_ret, diag_spy = spy_diag.compute_spy_diagnostics(spy_prices)

print("\nSPY efficiency diagnostics")
for key, value in diag_spy.items():
    print(f"{key:>8s} : {value: .4f}")
```

### 3. Side-by-Side Visualisation

The plots below compare a simulated baseline with empirical SPY behaviour. Treat them as a visual checklist before moving on to more elaborate tests.

```python
fig, axes = plt.subplots(2, 2, figsize=(10.0, 6.0))
(ax_rw_price, ax_spy_price), (ax_rw_hist, ax_spy_hist) = axes

ax_rw_price.plot(prices_rw)  # simulated price path
ax_rw_price.set_title("Random-walk price path")
ax_rw_price.set_ylabel("price level")
ax_rw_price.grid(True, alpha=0.3)

ax_spy_price.plot(spy_prices.index, spy_prices.values)
ax_spy_price.set_title("SPY end-of-day price path")
ax_spy_price.set_ylabel("price level")
ax_spy_price.grid(True, alpha=0.3)

ax_rw_hist.hist(log_ret_rw, bins=40, alpha=0.8, color="tab:blue")
ax_rw_hist.set_title("Random-walk log-returns")
ax_rw_hist.set_xlabel("log-return")
ax_rw_hist.grid(True, alpha=0.3)

ax_spy_hist.hist(spy_log_ret.values, bins=40, alpha=0.8, color="tab:orange")
ax_spy_hist.set_title("SPY daily log-returns")
ax_spy_hist.set_xlabel("log-return")
ax_spy_hist.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()
```

### Takeaways

- Comparing simulated and empirical diagnostics keeps your EMH baseline concrete.
- Even when markets are efficient in expectation, small autocorrelations and t-statistics appear purely from noise.
- Chapter 2’s later tests build on exactly these summary numbers, so keep this notebook handy as a reference.
